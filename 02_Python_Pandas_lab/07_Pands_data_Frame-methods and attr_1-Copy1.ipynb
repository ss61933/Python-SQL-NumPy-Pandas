# Apply method 

```python
ss["age"].unique()
```

```python
def salaryincr(age):
    if age<50:
        return "salary 500"
    else:
        return "salary 100"
```

```python
ss['age'].apply(salaryincr)
```

```python
ss['salary']=ss['age'].apply(salaryincr)
```

```python
ss
```

```python
def setnewsal(comm):
    if comm< 1000:
        return 100
    else:
        return 500
```

```python
ss['newsal']=ss['comm'].apply(setnewsal)
```

# Apply on multiple column

```python
def desc(row):
    # passs entirerow instead of column
    ID=row[0]
    salary=row[5]
    temp =str(ID) + ' ' + str(salary )
    return temp
```

```python
ss.apply(func=desc,axis='columns')
```

```python
ss['multicol']=ss.apply(func=desc,axis='columns')
# if we are passing entire row then follow this syntax
```

```python
ss
```
