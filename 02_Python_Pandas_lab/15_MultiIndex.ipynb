```python
#Unstack and stack the data
```

```python
pdata_f1=pdata.iloc[0:3,0:3]
```

```python
pdata_f1
```

```python
#Stacking covert columns to indexes #So normal columns are transposed to index 
```

```python
pdata_f1.stack()
```

```python
pdata_f12=pdata_f1.stack()
```

```python
pdata_f12.unstack()
```

```python
pdata_f12.head(6).unstack()
```

```python
pdata_f12.head(6).unstack(0) # So we are forcing to apply unstacking on Region only so region become column
```

```python
pdata_f12.head(6).unstack(1)    #Here state becomes columns 
```

```python
pdata_f12.head(6).unstack('Region') # We an pass column name as well 
```
