# Resampling

```python
#How many visiors we have for res1 in 2017?
```

```python
#Similarly getting count of records based on quarter,month,year etc is resapmling 
```

```python
pr1=pr.resample("A")
```

```python
pr1
```

```python
pr1.min()
```

```python
pr1.sum() #total visiors for each res
```

```python
pr1.mean()
```

```python
pr1.plot()
```

```python
pr1.mean().plot()
```

```python
#So above is example of time series analysis
```

```python
pr['2016-01'].sum()  #Sum of visitors for each restaurant for 1st month 2016
```

```python
#quarter basis analysis
```

```python
pr.resample("Q")
```

```python
pr1=pr.resample("Q")
```

```python
pr.resample("M")
```

```python
weekly=pr.resample("W")
```

```python
weekly.agg(["sum","mean"])
```

```python
weekly.agg(["sum","mean","max"])
```
