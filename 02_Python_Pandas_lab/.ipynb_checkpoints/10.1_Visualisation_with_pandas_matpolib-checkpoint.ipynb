# Data visualisation using matpotlib

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# which kind of style 
```

```python
plt.style.available
```

```python
#set the style
plt.style.use('bmh')
```

```python
x=np.linspace(-10,10,100)
y=np.sin(x)
plt.plot(x,y)
```

```python
plt.plot(x,np.cos(x))
```

```python
plt.plot(x,y,'--')
```

```python
#sub ploting concept 
# argument - rows,colunmns,which plot
```

```python
plt.subplot(2,1,1)
plt.plot(x,x)
```

```python
plt.subplot(2,1,2)
plt.plot(x,x**2)
```

```python
plt.subplot(1,1,1)
plt.plot(x,x**2)
```

```python
plt.plot(x,y,color='blue')
plt.plot(x,x,color="red")
```
