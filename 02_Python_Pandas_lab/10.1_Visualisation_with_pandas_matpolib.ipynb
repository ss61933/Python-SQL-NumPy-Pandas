```python
import numpy as np
```

```python
#set the style
plt.style.use('bmh')
```

```python
x=np.linspace(-10,10,100)
y=np.sin(x)
plt.plot(x,y)
```

```python
plt.plot(x,np.cos(x))
```

```python
plt.plot(x,y,'--')
```

```python
#sub ploting concept 
# argument - rows,colunmns,which plot
```

```python
plt.subplot(2,1,1)
plt.plot(x,x)
```

```python
plt.subplot(2,1,2)
plt.plot(x,x**2)
```

```python
plt.subplot(1,1,1)
plt.plot(x,x**2)
```

```python
plt.plot(x,y,color='blue',linestyle='-')
plt.plot(x,x,color="red",linestyle='--')
```

```python
#passing single argument for both 
plt.plot(x,y,':r')
```

```python
#change limit of x
plt.xlim(-5,15)
plt.ylim(-10,10)

plt.plot(x,y)
```

```python
plt.plot(x,x)
plt.axis([0,5,-4,6])
plt.title("y=x")
plt.xlabel("X label")
plt.ylabel("Y label")
```
